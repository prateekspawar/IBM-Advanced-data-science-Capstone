```python
import numpy as np
```

```python
X_train = df_train[df_train.columns[1:17]] 
Y_train = df_train['Confirmed']
X_test = df_test[df_test.columns[1:17]] 
Y_test = df_test['Confirmed']
X_train_np=X_train.to_numpy()
Y_train_np=Y_train.to_numpy()
X_test_np=X_test.to_numpy()
Y_test_np=Y_test.to_numpy()
X_train_np=np.reshape(X_train_np,(X_train_np.shape[0],X_train_np.shape[1]))
X_test_np=np.reshape(X_test_np,(X_test_np.shape[0],X_test_np.shape[1]))
Y_train_np=np.reshape(Y_train_np,(Y_train_np.shape[0],1))
Y_test_np=np.reshape(Y_test_np,(Y_test_np.shape[0],1))
```

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D,LeakyReLU,Dropout,InputLayer
from keras.initializers import glorot_normal,Zeros
from keras.regularizers import l2
import tensorflow as tf
from keras.metrics import mae
```

```python
kernal_ini=glorot_normal(seed=2)
bias_ini=Zeros()
l2_reg=l2(0.01)
l_re_alpha=0.01
```

Lets define a Deep learning model without regulalizers.

```python
model3 = Sequential()
   
#layer1
model3.add(Dense(16,use_bias=True,kernel_initializer=kernal_ini,bias_initializer=bias_ini,input_shape=(16,)))
#layer1_activation
model3.add(LeakyReLU(alpha=l_re_alpha))
# layer2
model3.add(Dense(32,use_bias=True,kernel_initializer=kernal_ini,bias_initializer=bias_ini))
#layer2_activation
model3.add(LeakyReLU(alpha=l_re_alpha))

#layer3
model3.add(Dense(32,use_bias=True,kernel_initializer=kernal_ini,bias_initializer=bias_ini))
#layer3_activation
model3.add(LeakyReLU(alpha=l_re_alpha))

#layer4
model3.add(Dense(16,use_bias=True,kernel_initializer=kernal_ini,bias_initializer=bias_ini))
#layer4_activation
model3.add(LeakyReLU(alpha=l_re_alpha))

#layer5
model3.add(Dense(8,use_bias=True,kernel_initializer=kernal_ini,bias_initializer=bias_ini))
#layer5_activation
model3.add(LeakyReLU(alpha=l_re_alpha))
#output
model3.add(Dense(1))
#compil model
model3.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_squared_error'])
```

```python
model3.summary()
```

```python
model3.fit(X_train_np,Y_train_np,epochs=500)
```

```python
from sklearn.metrics import mean_squared_error
y3_pred=model3.predict(X_test_np)
for i in range(len(y3_pred)):
    if(y3_pred[i]<0):
        y3_pred[i]=0
model_3_mse=mean_squared_error(Y_test_np, y3_pred)
model_3_mse
```

Lets define another Deep learning model including regularization and also dropout layers.

```python
model4 = Sequential()
   
#layer1
model4.add(Dense(16,use_bias=True,kernel_initializer=kernal_ini,bias_initializer=bias_ini,input_shape=(16,)))
#layer1_activation
model4.add(LeakyReLU(alpha=l_re_alpha))
# layer2
model4.add(Dense(32,use_bias=True,kernel_initializer=kernal_ini,bias_initializer=bias_ini,kernel_regularizer=l2_reg,activity_regularizer=l2_reg))
#layer2_activation
model4.add(LeakyReLU(alpha=l_re_alpha))
model4.add(Dropout(rate=0.2))
#layer3
model4.add(Dense(32,use_bias=True,kernel_initializer=kernal_ini,bias_initializer=bias_ini,kernel_regularizer=l2_reg,activity_regularizer=l2_reg))
#layer3_activation
model4.add(LeakyReLU(alpha=l_re_alpha))

#layer4
model4.add(Dense(16,use_bias=True,kernel_initializer=kernal_ini,bias_initializer=bias_ini,kernel_regularizer=l2_reg,activity_regularizer=l2_reg))
#layer4_activation
model4.add(LeakyReLU(alpha=l_re_alpha))
model4.add(Dropout(rate=0.2))
#layer5
model4.add(Dense(8,use_bias=True,kernel_initializer=kernal_ini,bias_initializer=bias_ini,kernel_regularizer=l2_reg,activity_regularizer=l2_reg))
#layer5_activation
model4.add(LeakyReLU(alpha=l_re_alpha))
model4.add(Dropout(rate=0.2))
#output
model4.add(Dense(1))
#compil model
model4.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_squared_error'])
```

```python
model4.fit(X_train_np,Y_train_np,epochs=500)
from sklearn.metrics import mean_squared_error
y4_pred=model4.predict(X_test_np)
for i in range(len(y4_pred)):
    if(y4_pred[i]<0):
        y4_pred[i]=0
model_4_mse=mean_squared_error(Y_test_np, y4_pred)
model_4_mse
```
